# üè† California House Price Prediction using Linear Regression

## Problem Statement:

You are working with a real estate analytics team that helps housing agencies and government planners understand how different factors affect property prices across California. Using data from the 1990 California Census, your task is to build a Linear Regression model that predicts the median house value in a given census block, based on numerical features such as median income, housing age, total rooms, population, and more.

The dataset includes over 20,000 records and covers diverse regions including areas near the ocean, bay, and inland zones. The model will serve as a simple, interpretable tool to estimate housing prices and identify the most influential factors driving them.

<hr>

- **Features:**
    - longitude, latitude (location coordinates)
    - housing_median_age (median age of houses in the district)
    - total_rooms, total_bedrooms (number of rooms and bedrooms)
    - population (total people in the district)
    - households (total number of households)
    - median_income (median income of residents)
    - ocean_proximity (categorical feature indicating the district‚Äôs proximity to the ocean)
<br><br>

- **Target:**
    - median_house_value (median house price in USD for the block)
<br><br>

- **Size:**
    - 20,640 entries
    - 9 numeric features, 1 categorical feature, 1 numeric target variable

This project focuses on the practical application of linear regression for real-world housing data, highlighting the strengths and limitations of linear models in spatial and economic contexts

## Preprocessing Data

### Handling missing values

The `total_bedrooms` column has **207 missing values**.
so filling these with median of `total_bedrooms`.

```python
median = housing["total_bedrooms"].median()
housing["total_bedrooms"] = housing["total_bedrooms"].fillna(median)

print(f"Missing values count:\n{housing.isnull().sum()}")
```

### Handling the capped values

- **Housing Median Age:** Bimodal distribution with a sharp spike at 50 ‚Äî might be capped (a max age ceiling).
- **Median House Value:** Long tail + suspicious spike at 500001 ‚Äî this suggests capping of target values

let's se how we will handle this

```python
housing["median_house_value"].value_counts().head()
```

```python
housing["housing_median_age"].value_counts().head()
```

**median_house_value:**
- That **500001** value appears **965 times**, making it a strong cap.
- Since it's the **target column** for our **linear regression**, keeping it would mislead the model ‚Äî it can't learn what values would‚Äôve been above 500001.

**housing_median_age:**
- Peaks at **52.0**, but it's **a feature, not the target**.
- There's no need to drop these ‚Äî we can just **leave the capped age values in**.
That spike is just telling us a lot of neighborhoods hit the age ceiling of the dataset

```python
housing = housing[housing["median_house_value"] < 500001]
housing["median_house_value"].value_counts().head()
```

### Handle the categorical Data

the feature `ocean_proximity` is categorical and we are doing linear regression so we need to convert this in numerical values. We will use **OneHot Encoding** to do that.

```python
housing["ocean_proximity"].value_counts()
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Categorical column
cat_features = ["ocean_proximity"]

# OneHotEncoder with correct keyword for newer sklearn
encoder = OneHotEncoder(drop='first', sparse_output=False)

# Create transformer
column_transformer = ColumnTransformer(
    transformers=[
        ("onehot", encoder, cat_features)
    ],
    remainder='passthrough'  # keep other columns
)

# Apply the transformation
housing_encoded = column_transformer.fit_transform(housing)

# 1. Get new one-hot encoded column names
encoded_cols = column_transformer.named_transformers_["onehot"].get_feature_names_out(cat_features)

# 2. Get numeric column names (i.e., the rest)
numeric_cols = housing.drop(columns=cat_features).columns.tolist()

# 3. Combine all columns
final_cols = list(encoded_cols) + numeric_cols

# 4. Create the cleaned-up DataFrame
housing_prepared = pd.DataFrame(housing_encoded, columns=final_cols)

# Reset index
housing_prepared.reset_index(drop=True, inplace=True)

# Sanity check
housing_prepared.head()
```

```python
housing_prepared.describe()
```

## Train Linear Regression

### Split data into train and test

we are going to do **stratified sampling** on `median_income` as it showed **high correlation** with `median_housing_value`.
<br>
`median_income` ‚Üí 0.688

If we randomly split the data, we might underrepresent some income ranges in either training or test set. Stratified sampling ensures all income levels are proportionally represented in both sets ‚Äî this gives our model better generalization and fairer evaluation.

```python
from sklearn.model_selection import StratifiedShuffleSplit

# Step 1: Create income category column in housing_prepared for stratification
housing_prepared["income_cat"] = pd.cut(
    housing_prepared["median_income"],
    bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
    labels=[1, 2, 3, 4, 5]
)

# Step 2: Stratified split on housing_prepared
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_idx, test_idx in split.split(housing_prepared, housing_prepared["income_cat"]):
    strat_train_set = housing_prepared.loc[train_idx].drop("income_cat", axis=1)
    strat_test_set = housing_prepared.loc[test_idx].drop("income_cat", axis=1)

# Step 3: Separate features and target
X_train = strat_train_set.drop("median_house_value", axis=1)
y_train = strat_train_set["median_house_value"]
X_test = strat_test_set.drop("median_house_value", axis=1)
y_test = strat_test_set["median_house_value"]

# Step 4: Confirm shapes
print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

### Feature Scaling

Since we‚Äôre doing **linear regression**, it‚Äôs a good to **scale the features** (especially continuous ones) so that they're on a similar scale. This helps the model converge faster and avoids giving undue importance to high-magnitude features.

```python
from sklearn.preprocessing import StandardScaler

# Identify numerical columns excluding one-hot encoded ones
one_hot_cols = [col for col in X_train.columns if col.startswith("ocean_proximity")]
numerical_cols = [col for col in X_train.columns if col not in one_hot_cols + ["median_house_value"]]

# Initialize scaler
scaler = StandardScaler()

# Fit and transform training data
X_train_scaled = X_train.copy()
X_train_scaled[numerical_cols] = scaler.fit_transform(X_train[numerical_cols])

# Transform test data using the same scaler
X_test_scaled = X_test.copy()
X_test_scaled[numerical_cols] = scaler.transform(X_test[numerical_cols])

# Confirm scaling
X_train_scaled.head()
```

### Training Linear Regression model

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train_scaled, y_train)

y_pred = lin_reg.predict(X_test_scaled)
```

```python
# Create comparison DataFrame
results = pd.DataFrame({
    "Median Income": X_test["median_income"].values,
    "Actual Price": y_test.values,
    "Predicted Price": y_pred.round(1)
})

# Show first 10 results
print(results.head(10))
```

### Evaluate Model

```python
# Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)

# Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# R¬≤ Score
r2 = r2_score(y_test, y_pred)

# Print all
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R¬≤ Score: {r2:.4f}")
```

- **MAE ($44,136):** On average, model's predictions are off by about 44k. This is the average absolute difference between actual and predicted house prices.
  
- **RMSE ($59,669):** A bit higher than MAE due to squaring large errors. This is more sensitive to big mistakes and gives a sense of typical prediction error magnitude.

  
- **R¬≤ Score (0.6156):** About **61.5% of the variation** in house values is explained by your model. For a simple linear model, that‚Äôs a decent baseline, especially for real-world housing data.

## Conclusion

In this project, we built a **Linear Regression model** to predict median house values across California using data from the 1990 Census. The model was trained on key numeric and categorical features, including **median income, housing age, population, and proximity to the ocean**.

**üîç Performance Summary:**

- **Mean Absolute Error (MAE):** 44,136
- **Root Mean Squared Error (RMSE):** 59,669
- **R¬≤ Score:** 0.6156

These metrics suggest that, on average, our predictions deviate from actual house values by around **$44,000**, and the model explains about **61.5% of the variance** in housing prices. While the model captures general trends (especially the strong correlation between income and price), the relatively high error values indicate that **housing prices are influenced by non-linear relationships** and complex interactions not fully captured by a simple linear model.

**üß† Takeaway:**

Linear regression provided a **simple and interpretable baseline**, helping identify influential features like **median_income**. However, to achieve more accurate predictions, especially in areas with extreme property values or unique geographic characteristics, **more advanced models** like **Random Forests or Gradient Boosting** could be explored in future iterations.
