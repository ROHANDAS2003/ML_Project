### Handling missing Values

Now that our dataset is cleaner, we need to **handle missing values** to avoid issues in analysis and model training.

<hr>

**Step 1: Understanding Missing Data**

Before filling in missing values, we check the **distribution of numerical columns** to decide the best approach.

```python
# Check distribution
plt.figure(figsize=(8, 2))
sns.histplot(studentsData['Attendance (%)'], bins=30, kde=True)
plt.title("Distribution of Attendance (%)")
plt.show()

plt.figure(figsize=(8, 2))
sns.histplot(studentsData['Assignments_Avg'], bins=30, kde=True)
plt.title("Distribution of Assignments_Avg")
plt.show()
```

**Why?** Checking the distribution helps us decide whether to use `mean`, `median`, or `another strategy` for filling missing values.

<hr>

**Step 2: Filling Missing Numerical Values**

Before filling let's **check the skewness** of these columns.
- If **skewness ≈ 0**, the data is **symmetrical**.
- If **skewness > 0**, it is **right-skewed** (long tail on the right).
- If **skewness < 0**, it is **left-skewed** (long tail on the left).

```python
from scipy.stats import skew

print("Attendance Skewness:", skew(studentsData['Attendance (%)'], nan_policy='omit'))
print("Assignments Avg Skewness:", skew(studentsData['Assignments_Avg'], nan_policy='omit'))
```

Since the skewness value is -0.038, it is **very close to 0**, meaning the Attendance (%) data is **nearly symmetrical** and not significantly skewed, so we can use mean to fill the missing values.

```python
# Since skewness is close to 0, use mean instead of median
studentsData['Attendance (%)'] = studentsData['Attendance (%)'].fillna(studentsData['Attendance (%)'].mean())
studentsData['Assignments_Avg'] = studentsData['Assignments_Avg'].fillna(studentsData['Assignments_Avg'].mean())
```

```python
plt.figure(figsize=(8, 2))
sns.histplot(studentsData['Attendance (%)'], bins=30, kde=True)
plt.title("Distribution of Attendance (%)")
plt.show()

plt.figure(figsize=(8, 2))
sns.histplot(studentsData['Assignments_Avg'], bins=30, kde=True)
plt.title("Distribution of Assignments_Avg")
plt.show()
```

**Why Mean?** Since the data is approximately normally distributed (skewness close to 0), the mean is a better choice as it preserves the natural distribution of the dataset.

<hr>

**Step 3: Filling Categorical Missing Values**

For **Parent_Education_Level**, we replace missing values with `"Unknown"` since it's categorical.

```python
studentsData['Parent_Education_Level'].value_counts()
```

```python
studentsData['Parent_Education_Level'] = studentsData['Parent_Education_Level'].fillna("Unknown",)
```

```python
studentsData['Parent_Education_Level'].value_counts()
```

**Step 4: Final Check for Missing Values**

```python
studentsData.isnull().sum()
```

All missing values are handled!

<hr>

**Instead of removing** rows or columns with missing values, we **filled them with appropriate values** to **prevent data loss** while maintaining data integrity.

- **Numerical Data:** Used `median` to fill missing values, ensuring robustness against outliers.
- **Categorical Data:** Used `"Unknown"` for missing values to retain all records without distorting categorical distributions.

This approach preserves all available data while making it clean and usable for analysis or machine learning models.

## Improving Grade System

The existing grading system showed inconsistencies when analyzed. The `Grade` column had a wide range of total scores for each grade.

By looking at the components which makes Total_scores we conclude tha the original Total_Score in the dataset did not properly reflect the weighted contributions of different assessment components.

```python
studentsData[["Midterm_Score","Assignments_Avg","Quizzes_Avg","Participation_Score","Projects_Score","Final_Score","Total_Score"]].head()
```

Clearly, the **Total Score does not seem to align well** with the input values.

```python
studentsData[["Attendance (%)", "Midterm_Score","Assignments_Avg","Quizzes_Avg","Participation_Score","Projects_Score","Final_Score","Total_Score"]].agg(["min", "max", "mean"])
```

**New Weight Distribution**

Based on a more realistic academic grading approach, we define the following weight distribution:  

| **Component**          | **Weight (%)** |
|------------------------|--------------|
| Attendance            | 5%           |
| Midterm Score        | 15%          |
| Assignments Avg      | 15%          |
| Quizzes Avg         | 10%          |
| Participation        | 5%           |
| Projects Score      | 20%          |
| Final Score         | 30%          |

This distribution ensures that **exams and projects hold significant weight**, while **participation and attendance contribute meaningfully** to the final grade.

```python
studentsData["Recalculated_Total"] = (
    (0.05 * studentsData["Attendance (%)"]) +
    (0.40 / 3 * (studentsData["Midterm_Score"] + studentsData["Assignments_Avg"] + studentsData["Quizzes_Avg"])) +
    (0.10 * studentsData["Projects_Score"]) +
    (0.40 * studentsData["Final_Score"]) +
    (0.05 * studentsData["Participation_Score"])
)

studentsData[["Total_Score", "Recalculated_Total"]].sample(10)
```

```python
studentsData[["Total_Score", "Recalculated_Total"]].agg(["min", "max", "mean"])
```

**Observations:**

- The **new scores are more evenly spread**.
- The **maximum possible score (90.09) is lower than before**, making grading adjustments necessary.

<hr>

Now as we are sure of this recalcualtion of total score lets use this.

```python
studentsData["Total_Score"] = studentsData["Recalculated_Total"]
studentsData.drop(columns="Recalculated_Total", inplace=True)
```

<hr>

**Fixing the Grade System**

To ensure a consistent grading scale, we redefined grades based on **Total Score** using a structured range:

- **A (0)** → 85 and above  
- **B (1)** → 75 - 84  
- **C (2)** → 65 - 74  
- **D (3)** → 55 - 64  
- **F (4)** → Below 55

We assigned these grades using a function that maps scores to the correct category.

```python
def assign_grade(score):
    if score >= 85:
        return "A"
    elif score >= 75:
        return "B"
    elif score >= 65:
        return "C"
    elif score >= 55:
        return "D"
    else:
        return "F"

studentsData["Grade"] = studentsData["Total_Score"].apply(assign_grade)

# Check if the new grades make sense
studentsData.groupby("Grade")["Total_Score"].agg(["min", "max", "mean"])
```

This **grading system** now **aligns well with real-world academic grading**.

```python
studentsData["Grade"].value_counts().sort_index()
```

### Results After Fixing  

After applying the new grading system:  

- `Grades` now have **distinct score ranges**:  
  - **A:** 85+  
  - **B:** 75 - 84  
  - **C:** 65 - 74  
  - **D:** 55 - 64  
  - **F:** Below 55

This structured grading system ensures **consistency**, making the **classification more reliable and interpretable**.


### Checking for Outliers

Since we've ensured proper data types, we should now focus on **outlier detection and treatment** before moving to scaling (standardization/normalization).

### Z-score

**Outliers** are data points that **significantly differ** from other observations in the dataset. These can affect the performance of machine learning models and distort analysis. To identify outliers, we used the `Z-score method`, which measures how far each data point is from the mean in terms of standard deviations.

**Steps Taken:**

- **Calculated Z-scores** for each numerical column.
- **Set a threshold of ±3**, meaning any value outside this range is considered an outlier.
- **Checked if any outliers exist** by counting values beyond this threshold.

```python
from scipy import stats
numerical_features = studentsData.select_dtypes(include=['number']).columns.tolist()

z_scores = stats.zscore(studentsData[numerical_features])
outliers_z = (abs(z_scores) > 3)  # Typically, anything beyond ±3 is an outlier

# Count outliers per column
outliers_z.sum(axis=0)
```

After applying the Z-score method, **no extreme outliers** were detected in the dataset. This suggests that our data is well-distributed without any significantly abnormal values.

Since **no outliers** were found, we do **not need to remove or transform** any values. This ensures that all data points remain intact for further processing.

```python
categorical_vars = [
    "Gender", "Extracurricular_Activities", "Parent_Education_Level", "Family_Income_Level", "Internet_Access_at_Home"
]

chi_square_results = []
for var in categorical_vars:
    contingency_table = pd.crosstab(studentsData[var], studentsData["Grade"])
    chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
    chi_square_results.append({
        "Variable": var.replace("_", " "),
        "Chi-Square Statistic": chi2,
        "P-value": p
    })

chi_square_results_df = pd.DataFrame(chi_square_results)
chi_square_results_df = chi_square_results_df.sort_values(by="P-value")

print(chi_square_results_df)
```

```python
plt.figure(figsize=(8, 6))
ax = sns.heatmap(
    chi_square_results_df.set_index("Variable"), 
    annot=True, cmap="Blues", fmt=".4f", linewidths=0.5, 
    linecolor="black", cbar=True
)
plt.title("Categorical Variable Significance in Relation to Grade", fontsize=14, fontweight="bold")

# Highlight significant values (P-value < 0.05)
for text in ax.texts:
    value = float(text.get_text())
    if value < 0.05: 
        text.set_color("red")  
        text.set_fontweight("bold")  

plt.xticks(rotation=0)  
plt.yticks(rotation=0)  
plt.show()
```

**Statistical Significance of Categorical Variables (Chi-Square Analysis)**

- **Most significant variable affecting grades:**
    - Internet access at home (p-value: 0.0009), meaning students with internet access are significantly more likely to achieve higher grades.

- **Other categorical variables (Extracurricular Activities, Income Level, Gender, Parent Education Level):**
    - These do not show a statistically significant impact on grades.

```python
anova_columns = [
    'Attendance (%)', 'Midterm_Score', 'Final_Score', 'Assignments_Avg', 
    'Quizzes_Avg', 'Participation_Score', 'Projects_Score'
]

anova_results = []

for col in anova_columns:
    groups = [studentsData[col][studentsData['Grade'] == grade].dropna() for grade in studentsData['Grade'].unique()]
    f_stat, p_value = stats.f_oneway(*groups)
    anova_results.append({'Variable': col, 'F-statistic': f_stat, 'P-value': p_value})

anova_df = pd.DataFrame(anova_results).sort_values(by="P-value")
print(anova_df)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Regression plot for Final Score vs Total Score
sns.regplot(
    x=studentsData["Final_Score"], 
    y=studentsData["Total_Score"], 
    scatter_kws={"s": 15, "alpha": 0.6, "color": "blue"}, 
    line_kws={"color": "red", "linewidth": 2}, 
    ax=axes[0]
)
axes[0].set_title("Regression: Total Score vs Final Score", fontsize=14, fontweight="bold", color="darkred")
axes[0].set_xlabel("Final Score", fontsize=12)
axes[0].set_ylabel("Total Score", fontsize=12)
axes[0].grid(True, linestyle="--", alpha=0.5)

# Regression plot for Midterm Score vs Total Score
sns.regplot(
    x=studentsData["Midterm_Score"], 
    y=studentsData["Total_Score"], 
    scatter_kws={"s": 15, "alpha": 0.6, "color": "green"}, 
    line_kws={"color": "red", "linewidth": 2}, 
    ax=axes[1]
)
axes[1].set_title("Regression: Total Score vs Midterm Score", fontsize=14, fontweight="bold", color="darkred")
axes[1].set_xlabel("Midterm Score", fontsize=12)
axes[1].set_ylabel("Total Score", fontsize=12)
axes[1].grid(True, linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

**Regression Analysis (Final Score vs. Total Score & Midterm Score vs. Total Score)**

- **Final Score vs. Total Score:**
Strong linear relationship (higher final scores generally mean higher total scores).

- **Midterm Score vs. Total Score:**
Weaker correlation compared to final scores.

## Conclusion

**The exploratory analysis reveals several key insights:**

- Internet Access at Home is a critical factor influencing student performance, with students having internet access significantly outperforming those without.

- Attendance plays a minor but positive role in total scores, though it is not a defining factor.

- Extracurricular Activities contribute positively to academic performance.

- Final Scores are the strongest predictor of overall Total Scores, while Midterm Scores have a weaker correlation.

- Study hours and stress levels do not have a direct, strong impact on grades, suggesting that quality of study and other external factors might be more influential.

Overall, academic success in this dataset appears to be most strongly influenced by access to resources (internet), engagement in extracurriculars, and consistent attendance rather than sheer study hours or stress levels.
